```python
pattern = r"(\d{2}/\d{2}/\d{2}),\s*(\d{1,2}:\d{2}\s*[ap]m)\s*-\s*([^:]+?):\s*(.+)"
```

```python
messages=re.split(pattern, data)[1:]
messages
```

```python
dates =  re.findall(pattern, data)
dates
```

```python
print("Messages:", len(messages))
print("Dates:", len(dates))
```

```python
matches = re.findall(pattern, data)
```

```python
df = pd.DataFrame(matches, columns=['date_str', 'time_str', 'user', 'message'])
 
print(df.head())
```

```python
df['date'] = pd.to_datetime(df['date_str'] + ' ' + df['time_str'], format='%d/%m/%y %I:%M %p')
```

```python
df['user'] = df['user'].str.strip()
df['message'] = df['message'].str.strip()
```

```python
df = df[['date', 'user', 'message', 'year', 'month', 'day', 'hour', 'minute']]
```
